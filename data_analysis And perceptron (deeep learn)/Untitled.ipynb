```python
x = np.array([1,2,3,4,5,6,8])
y=np.array([3,6,9,12,15,18,21])
iteration =100
```

```python
def gradientDescent(x,y,iteration,alpha=0.000012):
    n =  len(x)
    m_cur = c_cur = 0
    for _ in range(iteration):
        y_predicted = m_cur * x + c_cur
        dm = -(1/n) * sum(x *(y - y_predicted))
        dc = -(1/n)  * sum((y - y_predicted))
        m_cur =  m_cur - alpha*dm
        c_cur = c_cur -alpha * dc
        print("slope(m) = {} constant(c) = {} ,iteration {}".format(m_cur,c_cur,_))
        
        
        
        
    
```

```python
gradientDescent(x,y,iteration)
```
